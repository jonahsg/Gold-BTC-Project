```python
import pandas as pd
import numpy as np
```

```python
# Loading in Gold_Prices_Jan1979_to_Oct2020.csv to a Pandas DataFrame
gold_df = pd.read_csv('Gold_Prices_Jan1979_to_Oct2020.csv')
gold_df
```

```python
# Renaming 'Name' Column as 'Date'
# Renaming 'US Dollar' Column as 'Oz Gold/USD'
gold_df = gold_df.rename(columns = {'Name':'Date', 'US dollar':'Oz Gold/USD'})
```

```python
# Creating a new df, gold_usd_df, using only USD data
gold_usd_df = gold_df[['Date', 'Oz Gold/USD']]
gold_usd_df
```

```python
# Checking data types of gold_usd_df DataFrame
gold_usd_df.dtypes
```

```python
# Converting 'Date' column data type to datetime
gold_usd_df['Date']= pd.to_datetime(gold_usd_df['Date'])
```

```python
# Confirming 'Date' column data type is now datetime
gold_usd_df.dtypes
```
